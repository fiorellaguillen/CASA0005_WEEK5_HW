```python
%%sql

SELECT * FROM nyc_streets LIMIT 5
```

```python
%%sql

SELECT ST_AsText(geom)
FROM nyc_streets
WHERE NAME = 'Atlantic Commons';
```

```python
%%sql

SELECT * FROM nyc_neighborhoods LIMIT 5
```

```python
%%sql

SELECT NAME, BORONAME
FROM nyc_neighborhoods
WHERE ST_Intersects(geom, (SELECT geom FROM nyc_streets WHERE NAME = 'Atlantic Commons'))
```

```python
%%sql

SELECT * FROM nyc_streets LIMIT 5
```

```python
%%sql

SELECT NAME
FROM nyc_streets
WHERE ST_INTERSECTS(geom, (SELECT geom FROM nyc_streets WHERE NAME = 'Atlantic Commons'))
```

```python
%%sql

SELECT * FROM nyc_census_blocks LIMIT 5
```

```python
%%sql

SELECT sum(POPN_TOTAL) AS population
FROM nyc_census_blocks
WHERE ST_DWITHIN(
  geom,
  (SELECT geom FROM nyc_streets WHERE NAME = 'Atlantic Commons'),
  50
  );
```
